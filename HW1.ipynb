Point Charge

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline

plt.style.use('seaborn-colorblind')
```

```python
a = -5
b = -1*a
s = 0.1

x = np.arange(a,b+s,s)
y = np.arange(a,b+s,s)

X, Y = np.meshgrid(x,y)
```

```python
x1 = np.round(-1*3*s-s/2,4)
y1 = s/2

print('first charge at <',x1,',',y1,'>')
```

```python
epsilon = 8.85e-12
k = 1/(4*np.pi*epsilon)
q = 1e-9

E1x = k*q*(X-x1)/np.sqrt((X-x1)**2+(Y-y1)**2)**3
E1y = k*q*(Y-y1)/np.sqrt((X-x1)**2+(Y-y1)**2)**3

Enetx = E1x
Enety = E1y

skip = 6 ## reduce number of points to visualize
fig = plt.figure(figsize=(5,4),dpi=150)
plt.quiver(X[::skip,::skip],Y[::skip,::skip],Enetx[::skip,::skip],Enety[::skip,::skip])
plt.plot(x1,y1,'rx')
```

Electric Dipole

```python
a = -5
b = -1*a
s = 0.1

x = np.arange(a,b+s,s)
y = np.arange(a,b+s,s)
z = np.arange(a,b+s,s)

X, Y = np.meshgrid(x,y)
```

```python
x1 = np.round(-1*3*s-s/2,4)
y1 = s/2
x2 = np.round(3*s+s/2,4)
y2 = s/2

print('first charge at <',x1,',',y1,'>')
print('second charge at <',x2,',',y2,'>')
```

```python
epsilon = 8.85e-12
k = 1/(4*np.pi*epsilon)
q = 1e-9

E1x = k*q*(X-x1)/np.sqrt((X-x1)**2+(Y-y1)**2)**3
E1y = k*q*(Y-y1)/np.sqrt((X-x1)**2+(Y-y1)**2)**3

E2x = -k*q*(X-x2)/np.sqrt((X-x2)**2+(Y-y2)**2)**3
E2y = -k*q*(Y-y2)/np.sqrt((X-x2)**2+(Y-y2)**2)**3

Enetx = E1x + E2x
Enety = E1y + E2y

skip = 6 ## reduce number of points to visualize
fig = plt.figure(figsize=(5,4),dpi=150)
plt.quiver(X[::skip,::skip],Y[::skip,::skip],Enetx[::skip,::skip],Enety[::skip,::skip])
plt.plot(x1,y1,'rx')
plt.plot(x2,y2,'bx')
```

Electric Quadrapole

```python
a = -1.0
b = -1*a
s = 0.0151

x = np.arange(a,b+s,s)
y = np.arange(a,b+s,s)
z = np.arange(a,b+s,s)

X, Y = np.meshgrid(x,y)
```

```python
x1 = np.round(-1*3*s-s/2,4)
y1 = 0.105
x2 = np.round(3*s+s/2,4)
y2 = 0.105
x3 = np.round(-1*3*s-s/2,4)
y3 = -0.105
x4 = np.round(3*s+s/2,4)
y4 = -0.105

print('first charge at <',x1,',',y1,'>')
print('second charge at <',x2,',',y2,'>')
print('third charge at <',x3,',',y3,'>')
print('fourth charge at <',x4,',',y4,'>')
```

```python
epsilon = 8.85e-12
k = 1/(4*np.pi*epsilon)
q = 1e-9

E1x = k*q*(X-x1)/np.sqrt((X-x1)**2+(Y-y1)**2)**3
E1y = k*q*(Y-y1)/np.sqrt((X-x1)**2+(Y-y1)**2)**3

E2x = -k*q*(X-x2)/np.sqrt((X-x2)**2+(Y-y2)**2)**3
E2y = -k*q*(Y-y2)/np.sqrt((X-x2)**2+(Y-y2)**2)**3

E3x = -k*q*(X-x3)/np.sqrt((X-x3)**2+(Y-y3)**2)**3
E3y = -k*q*(Y-y3)/np.sqrt((X-x3)**2+(Y-y3)**2)**3

E4x = k*q*(X-x4)/np.sqrt((X-x4)**2+(Y-y4)**2)**3
E4y = k*q*(Y-y4)/np.sqrt((X-x4)**2+(Y-y4)**2)**3

Enetx = E1x + E2x + E3x + E4x
Enety = E1y + E2y + E3y + E4y

skip = 6 ## reduce number of points to visualize
fig = plt.figure(figsize=(5,4),dpi=150)
plt.quiver(X[::skip,::skip],Y[::skip,::skip],Enetx[::skip,::skip],Enety[::skip,::skip])
plt.plot(x1,y1,'rx')
plt.plot(x2,y2,'bx')
plt.plot(x3,y3,'bx')
plt.plot(x4,y4,'rx')
```
