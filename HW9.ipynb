```python
import sympy as sp               
import numpy as np               
import matplotlib.pyplot as plt  
from __future__ import division
```

```python
ang_I = np.arange(0,1001*np.pi/2000, np.pi/1000)
ang_T = np.arcsin(0.5*np.sin(ang_I))
E_0R_sq = 40000 # N^2/C^2
E_0T_sq = 17777 # N^2/C^2
v1 = 3e8 # m/s
v2 = 1.5e8 # m/s
eps_1 = 8.854e-12 # F/m
eps_2 = 3.5416e-11 # F/m
a = np.cos(ang_T)/np.cos(ang_I)
b = 2

E_0R_E_0I = (a-b)/(a+b)
E_0T_E_0I = 2/(a+b)

I_0R_I_0I = ((a-b)/(a+b))**2
I_0T_I_0I = a*b*(2/(a+b))**2
```

```python
fig = plt.figure(figsize=(12,12))
plt.plot(ang_I, E_0R_E_0I, label = 'Reflected amplitude')
plt.plot(ang_I, E_0T_E_0I, label = 'Transmitted amplitude')
plt.xlabel('Incident angle (radians)')
plt.ylabel('Amplitude of reflected or transmitted waves')
plt.title('Reflected and transmitted wave amplitudes vs. incident angle')
plt.annotate(s = 'Brewster angle', xy=(1.11,0), arrowprops=dict(arrowstyle='->'), xytext=(0.7,0.1))
plt.grid()
plt.legend()
plt.show()
```

This plot looks similar to the plot 9.16 in Griffiths. The Brewster's angle occurs at about $\theta_I = 1.11$ radians.
