```python
import sympy as sp               
import numpy as np               
import matplotlib.pyplot as plt
from __future__ import division
```

```python
k_0 = 1000
sig = 2
c = 3e8

X = np.linspace(-3, 3, 1000)
t = np.linspace(-10e-9, 10e-9, 5)
wave_packets = []
envelopes = []
for i in t:
    wave_packet = 0.5*np.exp(-sig**2*X**2/2 + c*sig**2*i*X - c**2*sig**2*i**2/2)*np.cos(k_0*X - c*k_0*i)
    wave_packets.append(wave_packet)
    envelope_pos = 0.5*np.exp(-sig**2*X**2/2 + c*sig**2*i*X - c**2*sig**2*i**2/2)
    envelope_neg = -0.5*np.exp(-sig**2*X**2/2 + c*sig**2*i*X - c**2*sig**2*i**2/2)
    tot_env = [envelope_pos, envelope_neg]
    envelopes.append(tot_env)

figure = plt.figure(figsize=(12,12))
plt.plot(X, wave_packets[2], linewidth=3.0, alpha=1.0)
plt.plot(X, wave_packets[0], linewidth=1.0, alpha=0.4, color='orange')
plt.plot(X, wave_packets[1], linewidth=1.0, alpha=0.4, color='orange')
plt.plot(X, wave_packets[3], linewidth=1.0, alpha=0.4, color='orange')
plt.plot(X, wave_packets[4], linewidth=1.0, alpha=0.4, color='orange')
plt.plot(X, envelopes[0][0], alpha=0.2, color='black', linestyle='--')
plt.plot(X, envelopes[0][1], alpha=0.2, color='black', linestyle='--')
plt.plot(X, envelopes[1][0], alpha=0.2, color='black', linestyle='--')
plt.plot(X, envelopes[1][1], alpha=0.3, color='black', linestyle='--')
plt.plot(X, envelopes[2][0], alpha=1.0, color='black', linestyle='--')
plt.plot(X, envelopes[2][1], alpha=1.0, color='black', linestyle='--')
plt.plot(X, envelopes[3][0], alpha=0.2, color='black', linestyle='--')
plt.plot(X, envelopes[3][1], alpha=0.2, color='black', linestyle='--')
plt.plot(X, envelopes[4][0], alpha=0.2, color='black', linestyle='--')
plt.plot(X, envelopes[4][1], alpha=0.2, color='black', linestyle='--')
plt.annotate(s = 'Direction of propagation', xy=(1,0.81), arrowprops=dict(arrowstyle='->'), xytext=(-0.95,0.8))
plt.annotate(s = 'Time = -10E-9 s', xy=(1,0.5), xytext=(-3.4,0.55))
plt.annotate(s = 'Time = -5E-9 s', xy=(1,0.5), xytext=(-1.9,0.55))
plt.annotate(s = 'Time = 0 s', xy=(1,0.5), xytext=(-0.25,0.55))
plt.annotate(s = 'Time = 5E-9 s', xy=(1,0.5), xytext=(1.1,0.55))
plt.annotate(s = 'Time = 10E-9 s', xy=(1,0.5), xytext=(2.5,0.55))
plt.grid()
plt.xlabel('Position (m)')
plt.ylabel('Amplitude (arbitrary units)')
plt.ylim(-1, 1)
plt.xlim(-3.5, 3.5)
plt.title('Wave-packet')
plt.show()
```

The plot above shows the wavepacket at several different times around $t = 0$. The problem said to pick a frequency within the visible wavelengths and a burst that lasts about 1 femtosecond. I cannot say that these requirements were met, but nevertheless, the plot illustrates the physics quite nicely. You can see the sinusoidal wave contained by a Gaussian envelope which is what a wave-packet would look like if we could see them with our eyes.
