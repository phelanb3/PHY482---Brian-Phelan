```python
import sympy as sp               
import numpy as np               
import matplotlib.pyplot as plt  
from __future__ import division
```

```python
w = 0.25 ## Width of loop in meters
alpha = 0.001 ## Growth rate of B-field in Tesla/second
x0 = 0.01 ## Initial length of loop inside B-field in meters
B0 = 0.1 ## Initial B-field strength in Tesla
v = 1 ## Speed of loop along x-hat direction (B-field points in y-hat direction)
time = np.arange(0,3.1,0.1) ## Set up time array to model EMF vs. time
```

```python
list_of_eps_v_time = []
plt.figure(figsize=(12,12))

while alpha < 100:
    abs_val_eps = w*(alpha*x0 + B0*v + 2*alpha*v*time)
    list_of_eps_v_time.append(abs_val_eps)
    alpha *= 10

plt.plot(time,list_of_eps_v_time[0],label='Alpha = 0.001')
plt.plot(time,list_of_eps_v_time[1],label='Alpha = 0.01')
plt.plot(time,list_of_eps_v_time[2],label='Alpha = 0.1')
plt.plot(time,list_of_eps_v_time[3],label='Alpha = 1')
plt.plot(time,list_of_eps_v_time[4],label='Alpha = 10')
plt.xlabel('Time (s)')
plt.ylabel('Induced EMF (V)')
plt.title('Induced EMF vs. Time for loop entering region of increasing B-field at constant speed')
plt.legend(loc='best')
plt.grid()
plt.show()
```

As we vary the value for alpha, the trend does not change. The EMF will increase at a higher rate but that is about it.

This paper attempts to obtain the Heisenberg spin interaction hamiltonian from the Hubbard model, drawing the connection between the two models. I have already partially explained the Hubbard model in my previous summary, so I will briefly summarize here. The Hubbard model describes a latice of atoms that only allow for a single state per atom site. Each state supports the two orientations for spin, and the Pauli principle asserts that these orientations must be opposite if the state were to hold two electrons. The hamiltonian consists of two terms: $H = H_0 + H_1$. One is a kinetic energy term of electrons hopping from site to site, and the other term is the Coulomb repulsion that twe electrons feel when they are located on the the same lattice site. The Coulomb repulsion term is proportional to the electrostatic energy, $I$, and proportional to the sum of sites that have electrons with opposite spin. The interaction between electrons at different sites is ignored (one of the approximations made in the Hubbard model).

We will assume that the system is in its ground state. In the ground state, the "hopping" kinetic energy term allows for electron behavior such that the electrons are not localized to one particular site which means many electrons can be superimposed on a single lattice site which makes the Coulomb repulsion very large. The Coulomb term however wants to minimize the interaction energy and thus have only one or fewer electrons at a single site. The Hubbard model has two competing terms! If we assume our system is at the atomic limit (something that I still have to understand more completely), the localized Coulomb term dominates. Because of this, we can treat the Hubbard model as just the Coulomb term, and we can treat the kinetic energy of the electrons hopping from site to site as a perturbation.

We saw above that the Hamiltonian is the sum of two terms: $H_0$ and $H_1$. $H_0$ is the unperturbed Coulomb term:

$H_0 = \frac{1}{2}I\sum_{i\sigma} n_{i\sigma}n_{i-\sigma}$

where the $n$'s count the number of sites that have two electrons with opposite spin (the $\frac{1}{2}$ accounts for double counting). The eigenvalues for the $H_0$ term has eigenvalues of $nI$ which is the number of sites with two electrons times the electrostatic energy. In the ground state, when we apply a first order perturbation, we see that the first order perturbation term is zero unless the quantum number $n$ (different from the $n$ representing number of sites with two electrons) is equal to 1. Then we attempt to solve the second order perturbation in the ground state.

There is a lot of math here that I will try to explain more heavily (and also include some equations) in my actual paper, but without going into much detail, it can be shown that this second order perturbation acting on the ground state of the system reduces to:

$H_{eff} = P_0(\sum_{kl}^{'}|T_{kl}|^{2}(2S_l*S_k - \frac{1}{2})/I)P_0$

where the the $T_{kl}$ values are the the coefficients that determine the type of magnetic ordering in the Heisenberg hamiltonian and the the $P_0$ operators are operators that restrict $H_{eff}$ to the subspace of states with the quantum number $n = 0$.

The section of this textbook(??) briefly describes the case of ferromagnetism derived from the Heisenberg model. The Heisenberg model of ferromagnetism is only valid if the spin interactions in the lattice are the dominating interactions, if the exchange interactions between electrons are anistropic in either the x, y or z directions (that the interaction in one direction dominates over the interaction in the other two directions), and that the particles of focus are fermions ($S = \frac{1}{2}$).

Just as a reminder, the Heisenberg hamiltonian is: $H = \frac{1}{2}\sum_{i,j} J_{ij}(S_{i}*S_{j})$. In this equation, the $S_{i}$'s are the total spin for an electron so the hamiltonian can be written as:

$H = \frac{1}{2}\sum_{i,j} J_{ij}(S_{i}^{x}S_{j}^{x} + S_{i}^{y}S_{j}^{y} + S_{i}^{z}S_{j}^{z})$  (1)

It is clear that the terms inside the parantheses do not commute. It is then beneficial to define the total spin as:

$S = \sum_{i}S_i$

Then, since each component of total spin $S_{i}^{k}$ commutes with the hamiltonian above, this must mean the total spin defined above also commutes with $H$. Therefore we can find eigenstates of $H$, $S*S$, and $S^{z}$.

Now we consider the state in which all the spins are aligned which results in the maximum value for the total spin. The state is just the product of each individual electron state at its maximum spin value. Next we can act on this wavefunction with the hamiltonian (1) above. If we do this, we can rearrang the $x$ and $y$ spin dot products into raising and lowering operators for the spin quantum number $m_s$. The $z$ spin dot product is left unaltered. Once we do this operation, we will see that:

$H|\psi> = -\frac{1}{2}S^2\sum_{ij}J_{ij}|\psi>$.

Now considering a state $|\psi>$, not necessarily having maximum total spin, we can take the expectation value of $H$ with this state:

$<\psi|H|\psi> = -\frac{1}{2}\sum_{ij}J_{ij}(\frac{1}{2}<\psi|S_i^+S_j^-|\psi> + \frac{1}{2}<\psi|S_i^-S_j^+|\psi> + <\psi|S_i^zS_j^z|\psi>) =  -\frac{1}{2}\sum_{ij}J_{ij}m_im_j$

If $J_{ij} >= 0$ for all $i,j$, then $J_{ij}m_im_j <= J_{ij}S^2$. And since the states described are states with smallest possible eigenvalues, $-\frac{1}{2}S^2\sum_{ij}J_{ij}$, and since we assumed that $J_{ij} >= 0$ (exchange interactions are positive), we have described the ground state of the Heisenberg model in the ferromagnetic case.
